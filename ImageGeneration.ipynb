```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ff = np.zeros([1024,1024])
```

```python
ff.shape
```

```python
np.random.random() - 0.5
```

```python
twopi = 2 * np.pi
k1 = twopi/256
k2 = twopi/64
k3 = twopi/128
k4 = twopi/32
```

```python
for n in range(1024):
    for m in range(1024):
        ff[n,m] = (100 * (np.sin(k1 * m) + np.cos(k2 * m)) * (np.sin(k3 * n) + np.cos(k4 * n)) +
                   20 * (np.random.random() - 0.5))
```

```python
fig, ax = plt.subplots()
ax.imshow(ff, origin='lower', cmap='gray')
```

```python
fnorm = np.max([np.abs(ff.max()), np.abs(ff.min())])
fnorm
```

```python
ffnorm = np.array(ff/fnorm*32768 + 32768, dtype = np.uint16)
im = Image.fromarray(ffnorm)
im.save("fig1.png")
```

```python
kx = 2 * np.pi/64 * np.cos(np.pi/6)
ky = 2 * np.pi/64 * np.sin(np.pi/6)
```

```python
gg = np.zeros([1024,1024])
for n in range(1024):
    for m in range(1024):
        gg[n,m] = (100 * np.sin(kx * m + ky * n) +
                   20 * (np.random.random() - 0.5))
```

```python
plt.imshow(gg, origin='lower', cmap='gray')
```

```python
gnorm = np.max([np.abs(gg.max()), np.abs(gg.min())])
gnorm
```

```python
ggnorm = np.array(gg/gnorm*32768 + 32768, dtype = np.uint16)
im = Image.fromarray(ggnorm)
im.save("fig2.png")
```
